Write a Python program that can do the following:

- You have $50

- You buy an item that is $15, that has a 3% tax

- Using the print()  Print how much money you have left, after purchasing the item.

```python
total = int(input("Enter total amount :"));
spending = int(input("Enter cost :"));
tax = int(input("Enter cost :"));
tax_amount = spending*(tax / 100)
expense = spending + tax_amount
balance = total - expense
print(balance)
```
