**Check if a number is odd/even**

```python
num = int(input("Enter a number: "))
if num % 2 == 0:
  print(f"{num} is even.")
else:
  print(f"{num} is odd.")
```

**Print multiplication table of n**

```python
num = int(input("Enter a number: "))
for i in range (1,11):
  prod = i * num
  print(f"{i} x {num} = {prod}")
```

**Reverse a string**

```python
s = input("Enter the string: ")
rev = s[::-1]
print(rev)

s = input("Enter the string: ")
rev = ""
for char in s:
    rev = char + rev     # add each character at the front
print(rev)
```

**Count vowels in a string**

```python
s = input("Enter the string: ")
vowels = "aeiouAEIOU"
count = 0

for char in s:
    if char in vowels:
        count += 1

print("Total vowels:", count)
```

**Sum of digits of a number**

```python
num = int(input("Enter a number: "))
total = 0

while num > 0:
    digit = num % 10
    total += digit
    num //= 10

print("Sum of digits:", total)
```

**Product of digits of a number**

```python
num = int(input("Enter a number: "))
total = 1

while num > 0:
    digit = num % 10
    total *= digit
    num //= 10

print("Product of digits:", total)
```

**Sum of odd and even digits of a number**

```python
num = int(input("Enter a number: "))
total_even = 0
total_odd = 0

while num > 0:
    digit = num % 10
    if num % 2 == 0:
      total_even += digit
    else:
      total_odd += digit
    num //= 10

print("Sum of even digits:", total_even)
print("Sum of odd digits:", total_odd)
```

**Palindrome**

```python
num = int(input("Enter a number: "))
temp = num
rev = 0

while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp //= 10

if rev == num:
    print("Palindrome")
else:
    print("Not Palindrome")
```

**Factorial**

```python
num = int(input("Enter a number: "))
fact = 1

for i in range(1, num + 1):
    fact *= i

print("Factorial:", fact)
```
