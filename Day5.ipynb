# **Dictionaries Assignment**

**Based on the dictionary:**

my_vehicle = {
    "model": "Ford",
    "make": "Explorer",
    "year": 2018,
    "mileage": 40000
}

**- Create a for loop to print all keys and values**

**- Create a new variable vehicle2, which is a copy of my_vehicle**

**- Add a new key 'number_of_tires' to the vehicle2 variable that is equal to 4**

**- Delete the mileage key and value from vehicle2**

**- Print just the keys from vehicle2**

```python
my_vehicle = { "model": "Ford", "make": "Explorer", "year": 2018, "mileage": 40000 }

for a, b in my_vehicle.items():
  print(f"{a} : {b}")
```

```python
my_vehicle = { "model": "Ford", "make": "Explorer", "year": 2018, "mileage": 40000 }
my_vehicle2 = my_vehicle
for a, b in my_vehicle2.items():
  print(f"{a} : {b}")
```

```python
my_vehicle = { "model": "Ford", "make": "Explorer", "year": 2018, "mileage": 40000 }
my_vehicle2 = my_vehicle
my_vehicle2["number_of_tires"] = 4
del my_vehicle2["mileage"]
for a, b in my_vehicle2.items():
  print(f"{a} : {b}")
```

```python
my_vehicle = { "model": "Ford", "make": "Explorer", "year": 2018, "mileage": 40000 }
my_vehicle2 = my_vehicle
for a in my_vehicle2:
  print(f"{a}")
```

```python
my_vehicle = { "model": "Ford", "make": "Explorer", "year": 2018, "mileage": 40000 }
my_vehicle2 = my_vehicle.copy()
print(my_vehicle2)
```

```python
my_vehicle = { "model": "Ford", "make": "Explorer", "year": 2018, "mileage": 40000 }
my_vehicle2 = my_vehicle.copy()
my_vehicle2.pop("mileage")
print(my_vehicle2)
```
