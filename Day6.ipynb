# **Functions Assignment**

**- Create a function that takes in 3 parameters(firstname, lastname, age) and returns a dictionary based on those values**

```python
def details(fname, lname, age):
  dict = {
      "fname": fname,
      "lname": lname,
      "age": age
  }
  return dict

user_dict = details(fname="Mark", lname="Wood", age=25)
print(user_dict)
```
